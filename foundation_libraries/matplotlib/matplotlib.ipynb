### 2D Line Plot

- Bivariate Analysis
- categorical -> numerical and numerical -> numerical
- Use case - Time series data

```python
# plotting a simple function
price = [48000,54000,57000,49000,47000,45000]
year = [2015,2016,2017,2018,2019,2020]

plt.plot(year,price)
```

```python
# from a pandas dataframe
batsman = pd.read_csv('sharma-kohli.csv')
batsman

plt.plot(batsman['index'], batsman['V Kohli'])
```

```python
# plotting multiple plots
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# labels title
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# colors(hex) and line(width and style) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'], linestyle='dashdot', color='#D9F10F', linewidth=4, marker='D', markersize=6)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# legend -> location
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='o',label='Rohit')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend() 
# plt.legend(loc='upper right') -> default loc is best

plt.grid()
```

```python
plt.barh(colors,children, color='grey', )
```

```python
df = pd.read_csv('batsman_season_record.csv')
df
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'],width=0.2,color='yellow')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='red')
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width=0.2,color='blue')

plt.xticks(np.arange(df.shape[0]), df['batsman'])

plt.show()
```

```python
x = np.arange(len(df['batsman']))  # position of each batsman on X-axis
width = 0.24  # width of each bar group

# Plot
plt.figure(figsize=(10, 6))

plt.bar(x - width, df['2015'], width, label='2015')
plt.bar(x, df['2016'], width, label='2016')
plt.bar(x + width, df['2017'], width, label='2017')

plt.xticks(x, df['batsman'], rotation=45)
plt.ylabel("Runs")
plt.title("Runs Scored by Batsmen (20152017)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Stacked Bar chart: stacking values on top of each other
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015') # 2015 bars must be placed on top of both -> sum of 2016 + 2017

plt.legend()
plt.show()
```

### Histogram 
- Univariate Analysis
- Numerical col
- Use case - Frequency Count

```python
df = pd.read_csv('vk.csv')
df
```

```python
plt.hist(df['batsman_runs'], bins=np.arange(0, 130, 10))
plt.show()
```

```python
# Load data
arr = np.load('big-array.npy')

# Create figure
plt.figure(figsize=(10, 6))

# Histogram with log scale
# Using log=True because:
# - The histogram counts vary greatly (large range in frequencies)
# - Small-frequency bins (like values 020) become almost invisible on a normal scale
# - log=True applies a logarithmic transformation to the y-axis
#   which spreads out small and large counts more evenly
# - This makes low-frequency bars visible instead of being squished to the bottom
plt.hist(
    arr,
    bins=[10, 20, 30, 40, 50, 60, 70],   # custom bins
    log=True,                           # log scale on y-axis
    edgecolor='black',                  # improve visibility
    alpha=0.7                           # slight transparency
)

# Labels and title
plt.xlabel("Value Range")
plt.ylabel("Frequency (log scale)")
plt.title("Histogram of Array Values (Logarithmic Y-Scale)")

# Improve layout
plt.tight_layout()
plt.show()
```
