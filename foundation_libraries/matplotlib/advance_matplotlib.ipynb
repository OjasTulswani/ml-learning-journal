```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Colored Sctterplots


```python
iris = pd.read_csv('iris.csv')
iris.sample(6)
```

```python
iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris.sample(5)
# replace() by default returns a new Series with the replacements
# assign that new Series back to iris['Species'] using assignment op.
```

```python
plt.scatter(
    iris['SepalLengthCm'],  # x-axis values: Sepal length of each flower
    iris['PetalLengthCm'],  # y-axis values: Petal length of each flower
    c=iris['Species'],      # color of each point based on species (0,1,2)
    cmap='jet',             # color map to use for coloring points
    alpha=0.7               # transparency of points (0=transparent, 1=opaque)
)

plt.xlabel('Sepal Length')  # Label for the x-axis
plt.ylabel('Petal Length')  # Label for the y-axis
plt.colorbar()              # Adds a colorbar to show the mapping from numbers (0,1,2) to colors
```

### Annotations


```python
batter = pd.read_csv('batter.csv')
batter.shape
```

```python
df = batter.head(100).sample(25, random_state=5)
```

```python
df
```

```python
plt.figure(figsize=(18,10))
plt.scatter(df['avg'], df['strike_rate'], s=df['runs'])

for i in range(df.shape[0]):
    plt.text(df['avg'].values[i], df['strike_rate'].values[i], df['batter'].values[i], ha='left', va='bottom', rotation=15)
```

### 3D Surface Plots

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

#np.meshgrid turns two 1D lists of x and y values into two 2D grids that represent all combinations of those x and y values
xx , yy = np.meshgrid(x, y)
xx.shape
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p) #
```

### Contour 3D to 2D

```python
ig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```
