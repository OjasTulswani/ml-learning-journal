## 1. NumPy (Numerical Python)
*The foundation of numerical computing in Python.*

### A. Creating Arrays
```python
import numpy as np

# From list
arr = np.array([1, 2, 3]) 

# Placeholders
np.zeros((3,3))    # 0s matrix
np.ones((2,5))     # 1s matrix
np.full((2,2), 7)  # Matrix of 7s
np.eye(3)          # Identity matrix

# Ranges
np.arange(0, 10, 2)     # [0, 2, 4, 6, 8] (Start, Stop, Step)
np.linspace(0, 1, 5)    # [0.0, 0.25, 0.5, 0.75, 1.0] (5 equal parts)

# Random (Crucial for weights initialization)
np.random.rand(3,2)     # Uniform distribution [0,1)
np.random.randn(3,2)    # Standard Normal distribution (Gaussian)
np.random.randint(1,100, 5) # 5 random integers
```

### B. Vector vs. Matrix vs. Tensor
*   **Vector (1D):** Shape `(n,)`. A list of numbers.
*   **Matrix (2D):** Shape `(n, m)`. Rows and columns (e.g., an Excel sheet).
*   **Tensor (3D+):** Shape `(n, m, k, ...)`. E.g., An image is `(Height, Width, RGB Channels)`.

### C. Array Reshaping
*Crucial for ML input formatting (Scikit-Learn expects 2D inputs).*
```python
arr = np.arange(10) # Shape (10,)

# Reshape to 2D matrix
arr.reshape(2, 5)   

# The "Magic" -1 (Auto-calculate dimension)
arr.reshape(-1, 1)  # Converts 1D vector to Column Vector (10, 1)
arr.reshape(1, -1)  # Converts 1D vector to Row Vector (1, 10)

# Flattening
arr.flatten()       # Returns copy (1D)
arr.ravel()         # Returns view (faster, 1D)
```

### D. Fancy Indexing & `np.where`
```python
arr = np.array([10, 20, 30, 40, 50])

# Boolean Masking
mask = arr > 30 
print(arr[mask])    # [40, 50]

# np.where (Vectorized If-Else)
# Syntax: np.where(condition, value_if_true, value_if_false)
np.where(arr > 30, 1, 0)  # Output: [0, 0, 0, 1, 1] (Great for binary classification labeling)
```

### E. Important Methods for ML
1.  **Statistics:** `np.mean()`, `np.median()`, `np.std()`, `np.var()`.
2.  **Linear Algebra:**
    *   `np.dot(a, b)` or `a @ b`: Matrix multiplication.
    *   `arr.T`: Transpose.
3.  **Finding Elements:**
    *   `np.argmax()` / `np.argmin()`: Returns **index** of max/min value (Used to convert Softmax probability to class label).
    *   `np.unique()`: Finds unique classes in a dataset.
4.  **Math:** `np.exp()` (Sigmoid/Softmax functions), `np.log()` (Log Loss).

## 4. Seaborn
*High-level interface for Matplotlib. Makes stats plots beautiful.*

### A. The "Axes-Level" vs. "Figure-Level" Distinction
This is the most confusing part of Seaborn.

#### 1. Axes-Level Functions
*   **What they do:** Draw onto a single matplotlib Axes (subplot).
*   **Return:** `matplotlib.pyplot.Axes` object.
*   **Usage:** Can be placed inside a Matplotlib subplot grid.
*   **Examples:** `sns.scatterplot`, `sns.barplot`, `sns.boxplot`, `sns.heatmap`.
*   **Arguments:** Accepts `ax=...` argument.

#### 2. Figure-Level Functions
*   **What they do:** Create a new figure entirely. They manage their own subplots (faceting).
*   **Return:** `Seaborn.FacetGrid` object.
*   **Usage:** Good for creating multiple subplots automatically based on a column (`col='Region'`).
*   **Examples:** 
    *   `sns.relplot` (handles scatter/line)
    *   `sns.displot` (handles hist/kde)
    *   `sns.catplot` (handles bar/box/violin)
*   **Arguments:** Does **NOT** accept `ax=...`. You control size via `height=` and `aspect=`.

### B. Graph Use Cases & Differences

| Graph Type | Function | Use Case & Nuance |
| :--- | :--- | :--- |
| **Heatmap** | `heatmap` | **Matrix visualization**. Essential for Confusion Matrices and Correlation Matrices (`df.corr()`). |
| **Violin Plot** | `violinplot` | **Boxplot + KDE**. Shows the summary stats (like box) AND the probability density (shape/fatness) of data. |
| **Count Plot** | `countplot` | Like a Histogram, but for **Categorical** data. Counts occurrences of strings/labels. |
| **Pair Plot** | `pairplot` | **EDA Powerhouse**. Plots pairwise relationships for the *entire* dataset at once. |
| **KDE Plot** | `kdeplot` | **Kernel Density Estimate**. A smooth version of a histogram. Good for visualizing probability distributions. |

### C. Quick Syntax Example (Figure vs Axes)

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Axes Level (Integration with plt)
fig, ax = plt.subplots(figsize=(5,5))
sns.boxplot(data=df, x='Category', y='Value', ax=ax) # We tell it where to sit

# 2. Figure Level (Auto-generates figure)
sns.relplot(data=df, x='Height', y='Weight', hue='Gender', col='Country') 
# 'col' automatically creates subplots for each country
```

# ðŸ Seaborn API Structure: Figure vs. Axes Level

## ðŸ”‘ The Golden Rule
1.  **Figure-Level:** Creates a **NEW** figure/window. Good for easy subplots (faceting) using `col=` and `row=` arguments. Returns a `FacetGrid`.
2.  **Axes-Level:** Draws onto an **EXISTING** specific plot (`ax=...`). Good for custom layouts with Matplotlib. Returns a `matplotlib.axes`.

## 3. Important Exceptions & Argument Notes

When you are coding, you might get an error saying `unexpected keyword argument 'hue'`. Here is why:

### 1. `sns.regplot()`
*   **Problem:** Does not support `hue`.
*   **Why:** It calculates specific regression statistics that are hard to stack on one axes object simply.
*   **Solution:** Use `sns.lmplot()`.

### 2. `sns.heatmap()`
*   **Problem:** Does not support `hue`.
*   **Why:** A heatmap uses color to represent the magnitude of the Z-axis (the value in the cell). Adding `hue` (another color dimension) would make the graph unreadable.

### 3. `sns.displot()` vs `sns.histplot()`
*   `displot` is **Figure-Level** (can create subplots with `col=`).
*   `histplot` is **Axes-Level** (goes onto a specific single plot).
*   *Note:* The old function `sns.distplot` (deprecated) did **not** support `hue` well. Always use `histplot` or `displot` now.

### 4. `plt` (Matplotlib standard functions)
*   Standard Matplotlib functions (`plt.plot`, `plt.scatter`, `plt.bar`) **DO NOT** support the `hue` argument.
*   **Workaround:** You have to loop through your data groups and plot them manually or color them using a mapped list of colors. (This is why we prefer Seaborn for ML exploration!)

## 4. Quick Summary for ML Projects

1.  **Check Target Distribution (Classification):** `sns.countplot(x='Target')` (Check for class imbalance).
2.  **Check Target Distribution (Regression):** `sns.histplot(x='Target', kde=True)` (Check for skewness).
3.  **Check Outliers:** `sns.boxplot(x='Feature')`.
4.  **Check Correlations:** `sns.heatmap(df.corr(), annot=True)`.
5.  **Check Feature vs Target:**
    *   If Target is Num: `sns.scatterplot(x='Feature', y='Target')`.
    *   If Target is Cat: `sns.boxplot(x='Target', y='Feature')` (Does this feature separate the classes well?).
