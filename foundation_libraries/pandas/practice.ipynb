```python
import pandas as pd
```

```python
data = pd.read_csv('deliveries.csv') #read the CSV file 'matches.csv', pandas by default make tables as dataframe
print(type(data)) # confirm that it is a of type dataframe 
print('\n')
print(data.dtypes) # the data types of each column in the dataframe
```

```python
data.head(2)
```

```python
# Ex-1 find out top 10 batsman of all the seasons who make most runs

batsman = data.groupby('batsman')
batsman['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
# Ex-2 find the batsman who made most fours
mask = data['batsman_runs'] == 4
new = data[mask].groupby('batsman')['batsman_runs']
new.count().sort_values(ascending=False).head()

```

```python
# ex 3 - create function to find a batsman scored max run against which team
def run_scored(batsman_name):
    data_set = data[data['batsman'] == batsman_name]
    return data_set.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head()

run_scored('V Kohli')
```

```python
# ex -4 Find the most destructive death over batsman in the history of IPL
# Strike rate = (No. of runs / no. of balls) * 100
# minimun batsman 200 balls in 16 - 20 overs

```

```python
death_over = data[data['over'] > 15]

batsmans = death_over.groupby('batsman')['batsman_runs'].count()
mask = batsmans > 200
batsmans_list = batsmans[mask].index.tolist()

result = death_over[death_over['batsman'].isin(batsmans_list)].groupby('batsman')

runs = result['batsman_runs'].sum()
balls = result['batsman_runs'].count()

outcome = (runs / balls) * 100

outcome.sort_values(ascending=False).head()
```
