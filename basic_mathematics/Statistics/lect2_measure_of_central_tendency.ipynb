```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('iris.csv')
data.head()
```

```python
data.rename(columns={
    'SepalLengthCm' : 'SL',
    'SepalWidthCm' : 'SW',
    'PetalLengthCm' : 'PL',
    'PetalWidthCm' : 'PW',
},
inplace=True)
```

```python
iris_setosa = data[data['Species'] == 'Iris-setosa']
iris_versicolor = data[data['Species'] == 'Iris-versicolor']
iris_virginica = data[data['Species'] == 'Iris-virginica']
```

### MEAN

```python
 # Sepal length
print('Setosa', np.mean(iris_setosa['SL'])) 
print('Versicolor', np.mean(iris_versicolor['SL'])) 
print('Virgnica', np.mean(iris_virginica['SL'])) 
```

```python
 # Sepal width
print('Setosa', np.mean(iris_setosa['SW'])) 
print('Versicolor', np.mean(iris_versicolor['SW'])) 
print('Virgnica', np.mean(iris_virginica['SW'])) 
```

```python
 # Petal length
print('Setosa', np.mean(iris_setosa['PL'])) 
print('Versicolor', np.mean(iris_versicolor['PL'])) 
print('Virgnica', np.mean(iris_virginica['PL'])) 
```

```python
 # Petal width
print('Setosa', np.mean(iris_setosa['PW'])) 
print('Versicolor', np.mean(iris_versicolor['PW'])) 
print('Virgnica', np.mean(iris_virginica['PW'])) 
```

## Problems with the Mean

The mean is not always the best measure of central tendency because it can be affected by certain data characteristics.

**a) Sensitive to Outliers:**
Extreme values can pull the mean away from the typical data.

**b) Not Suitable for Skewed Data:**
When data is lopsided(unevenly distributed), the mean may not represent the “middle” value well.

**c) May Not Be an Actual Data Value:**
The mean can be a number that doesn’t exist in the dataset.

**d) Not Suitable for Qualitative Data:**
The mean only works with numerical data, not categories or labels.

**e) Affected by Missing or Incorrect Data:**
Wrong or missing values can significantly change the mean.


```python
# range of setosa petal length is between 1 to 2
plt.hist(iris_setosa['PL'])
```

```python
# now we are adding n outlier of 50 at the end
print(np.mean(np.append(iris_setosa['PL'], 50)))

# here we can see that the mean is shiftes towards one side on existing of an outlier, i.e its go beyound the range of 1 to 2 
```

### Median and Outliers
 Median is a good measure of central tendency as long as extreme values (outliers) are not too many. It is less affected by outliers compared to the mean.

```python
 # Petal length
print('Setosa', np.median(iris_setosa['PL'])) 
print('Versicolor', np.median(iris_versicolor['PL'])) 
print('Virgnica', np.median(iris_virginica['PL'])) 
```

```python
# now we are adding n outlier of 50 at the end
print(np.median(np.append(iris_setosa['PL'], 50)))
```

### Mode

```python
from scipy import stats

print(stats.mode(iris_setosa['PL']))
print(stats.mode(iris_versicolor['PL']))
print(stats.mode(iris_virginica['PL']))
```

# Interpreting Central Tendency

| Mean vs Median vs Mode       | Interpretation                       
|------------------------------|--------------------------------------|
| Mean ≈ Median ≈ Mode         | Symmetrical distribution             |
| Mean > Median > Mode         | Positively skewed (tail to right)    |
| Mean < Median < Mode         | Negatively skewed (tail to left)     |
| Mean ≠ Median ≠ Mode         | Skewed or unusual distribution       |

**Tip:**  
- Compare mean and median to quickly spot skewness.  
- Use mode to identify the most common value.
