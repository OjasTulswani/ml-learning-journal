### Fetching with conditions

```python
mask = data['city'] == 'Indore'
data[mask]

```

```python
# fuction to get matches by city
def get_city_match_count(city):
    mask=data['city'] == city
    return data[mask].shape[0]
```

```python
print('No. of matches in Hyderabad: ', get_city_match_count('Hyderabad'))
```

```python
# multi condition
mask1 = data['city']=='Hyderabad'
mask2 = data['date'] >= '2010-01-01'
print('Matches played after 2010 in Hyderabad:', data[mask1 & mask2].shape[0])
```
