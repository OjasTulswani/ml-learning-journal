### MERGE


```python
# merge() function is used to combine two DataFrames, similar to SQL joins

# for understanding merge we find the orange cup holder for the perticular season
delivery = pd.read_csv('deliveries.csv')
```

```python
data.head(1)
```

```python
delivery.head(1)
```

```python
mergered_dataset = delivery.merge(data, left_on='match_id', right_on='id')
```

```python
mergered_dataset.groupby(['season', 'batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset='season', keep='first').sort_values('season')
# reset_index -> to move the index back into a normal column
```

### PIVOT TABLE AND PIVOT


```python
# Pivot: reshaping a table
# - Converts unique values from a column into separate new columns
# - Rearranges the data without performing calculations
# - Requires each (index, column) pair to be unique


# Pivot table:
# - Does everything pivot can do
# - Can summarize, aggregate, combine data, and handle duplicates

```

```python
# create a pivot table so that for each team we can determine that how much fours or sixs in a particular table
```

```python
mask = delivery['batsman_runs'] == 6
six = delivery[mask]


pt = six.pivot_table(index='over', columns='batting_team', values='batsman_runs', aggfunc = 'count')
pt
```

```python
import seaborn as sns
```

```python
sns.heatmap(pt)
```
